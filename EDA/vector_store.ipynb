```python
from langchain_community.document_loaders.csv_loader import CSVLoader

loader_file = CSVLoader(
						file_path='/content/drive/MyDrive/workspace/kaggle&github/~09.30_빅콘테스트/data/제주_관광_데이터/sh_craw_concat.csv',
                        encoding='utf-8' # cp949 or utf-8
)

files = loader_file.load()
# 시험용으로 10개만 추출
files = files[:10]

print(len(files))
```

```python
files[0]
```

```python
!pip install tiktoken


from langchain_text_splitters import TokenTextSplitter

text_splitter = TokenTextSplitter(
    chunk_size=200,  # 청크 크기를 100으로 설정
    chunk_overlap=0,  # 청크 간 중복을 0으로 설정
    add_start_index=True # 청크의 시작 인덱스 표기
)

# csv파일을 청크로 분할
splits = text_splitter.split_documents(files) # document 형식으로 불러온 file을 청크로 분할
print("분할된 csv의 첫번재 청크 : ",splits[0])
print("분할된 청크의 개수 : ", len(splits))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

modelPath = "jhgan/ko-sroberta-multitask"
model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings': True} # 벡터간의 cosine similarity 계산에서 더 유리하다.

embeddings_model = HuggingFaceEmbeddings(
    model_name=modelPath,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores.utils import DistanceStrategy

vectorstore = FAISS.from_documents(documents = splits,
                                   embedding = embeddings_model,
                                   distance_strategy = DistanceStrategy.COSINE
                                  )
vectorstore
```

prompft template
- 음식점에 대해 추천하는 것이기 때문에 질문의 구성을 생각해야함.
- 역할 : 음식점 추천 어시스턴트 / 해당하는 조건의 음식점을 추천하고, 그 테마와 맞는 관광지도 동시에 추천

```python
# 챗봇의 역할 정의
## 1. 음식점의 디테일, 추가적인 관광지 추천
## 2. context에 어긋나면 다시 물어보기
## 3. question을 넣고싶으면 Question : {question} 추가

template = """You are a restaurant recommendation assistant. Use the following pieces of context to recommend a restaurant that matches the user's preferences.

The answer should include details about the restaurant, the menu, who typically visits the restaurant, and any notable nearby landmarks or views. answers also include images of restaurant and landmarks(views).
If you don't know the answer, just say that you don't know, don't try to make up an answer.
Always answer in Korean.

If the question is not related to the context, politely respond that you are tuned to only answer questions that are related to the context.

if any information such as companion, budget or view is missing from the user's question, use the context only provided and recommen based on general preferences of giving informations or similar query.
If context does not provide sufficient information, make an educated guess based on common preferences.

Helpful Answer:
- Who: {companion}
- Menu: {menu}
- Budget: {budget}
- View or nearby landmarks: {view}

Context provided:
{context}


Answer includes :
- A brief description of the restaurant.
- Main medu offered at the restaurant.
- who typically visits the restaurant.(e.g., family, couples, friends, parents, with kids, tourists..)
- Any notable nearby landmarks or views.
- images of restaurant and landmarks(views).
- relavant information of other restaurant.(recommend another restaurants)

"""

QA_CHAIN_PROMPT = PromptTemplate.from_template(template)
```

```python
# Import the RetrievalQA class from the langchain module
from langchain.chains import RetrievalQA

# 리트리버 체인 설정
# Create a RetrievalQA instance with specified components
chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={
        "prompt": QA_CHAIN_PROMPT,
        "document_variable_name": "context"  # 문서 내용을 context로 매핑
    }
)
```

```python
context = '임베딩된 맛집 정보'
question = '20대가 많이 찾는 맛집은 어디인가요?'
companion = '20대'
budget = '10만원'
view = '바다'
menu = '장어'
```

```python
docs = retriever.get_relevant_documents(question)

# 검색된 문서 개수 확인
print(f"검색된 문서 개수: {len(docs)}")

# 첫 번째 문서의 내용 확인
if len(docs) > 0:
    print("첫 번째 문서 내용:", docs[0].page_content)
else:
    print("문서를 찾지 못했습니다.")
```

```python
from langchain.chains import LLMChain
headers = {'Autorization' : 'Bearer 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'}

chain = LLMChain(
    llm=llm,
    prompt=QA_CHAIN_PROMPT
)

formatted_query = QA_CHAIN_PROMPT.format(
    context = context,
    question = question,
    companion=companion if companion != '' else "일반적인 방문객",
    menu=menu if menu != '' else "주요 메뉴 정보 없음",
    budget=budget if budget != '' else "가격 정보 없음",
    view=view if view != '' else "뷰 정보 없음"
)

# 체인 실행 - run()으로 인해 바로 실행됨
## 토큰도 같이 넣어주기
result = chain.run({
    "question" : question,
    "companion": companion,
    "menu": menu,
    "budget": budget,
    "view": view,
    "context": "임베딩된 맛집 정보",  # 필요한 문서 또는 정보
}, headers=headers)

# 결과 출력
print("답변: ", result)
```
