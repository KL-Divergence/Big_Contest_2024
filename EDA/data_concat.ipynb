#리뷰에 대한 그룹핑
- 하나의 관광지에 대해서 리뷰를 통합한 후 키워드를 칼럼으로 만들기
- 연인, 커플, 남친/여친 등을 찾아봤는데 그렇게 많지 않았음.
- 이런거는 어쩔수없이 llm의 pretrain emb model에 맡겨야할듯?


```python
# 다른 키워드 찾아보기
score_tot[score_tot['MOP_CN'] == '커플']
```

그래서 일단은 리뷰를 나누기 위해 상위 count 상위 20개(약 91%를 차지하는)단어에 대해서

```python
import re
def word_count(word, reviews) :
    regex = re.escape(word)
    return sum(len(re.findall(regex, str(review))) for review in reviews)
```

```python
grouped = score_tot.groupby('TRRSRT_NM')['MOP_CN'].apply(list).reset_index()
grouped
```

```python
word_to_search = ['아이', '친절', '볼거리', '사진', '시간', '구경', '생각', '제주도', '방문',
                  '최고', '제주', '힐링', '추천', '산책', '추억', '날씨', '아기', '부모']


for word in word_to_search :
    grouped[word] = grouped['MOP_CN'].apply(lambda x : word_count(word, x))


print(grouped[['TRRSRT_NM'] + word_to_search].head())
```
